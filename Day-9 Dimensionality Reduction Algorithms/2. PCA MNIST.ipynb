## PCA MNIST Example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.decomposition import PCA
```

## 784 Features --> 100 Features --> 784 Features

```python
data = np.load('../../Datasets/mnist/mnist_train_small.npy')

X = data[:, 1:]
```

```python
model = PCA(n_components=100)
```

```python
model.fit(X)
```

```python
model.components_.shape
```

```python
X.shape
```

```python
model.components_.shape
```

```python
X_mod = np.dot(X, model.components_.T)
```

```python
X_mod.shape
```

```python
X_regen = np.dot(X_mod, model.components_)
```

```python
plt.imshow(X_regen[0].reshape(28, 28), cmap='gray')
```

```python
plt.imshow(X[0].reshape(28, 28), cmap='gray')
```

```python
X[:5]
```

```python
X_regen[:5]
```

# Visualizing in 2D

```python
X, y = data[:, 1:], data[:, 0]
```

```python
y.shape
```

```python
model = PCA(n_components=2)
```

```python
model.fit(X)
```

```python
model.components_.shape
```

```python
X_mod = np.dot(X, model.components_.T)
```

```python
X_mod.shape
```

```python
plt.scatter(X_mod[:, 0], X_mod[:, 1], c=y)
```

```python
import seaborn as sns
```

```python
selected = y < 3  # Selecting from 0 to 2
```

```python
y_selected = y[selected]
X_selected = X_mod[selected]
```

```python
plt.scatter(X_selected[:, 0], X_selected[:, 1], c=y_selected)
```

```python
sns.scatterplot(X_selected[:, 0], X_selected[:, 1], hue=y_selected)
```
